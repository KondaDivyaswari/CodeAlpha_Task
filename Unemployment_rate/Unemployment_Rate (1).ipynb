```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Unemployment/Unemployment_Rate_upto_11_2020.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Rename for convenience
df = df.rename(columns={
    'Estimated Unemployment Rate (%)': 'UnemploymentRate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'LabourParticipationRate'
})

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])

# Preview
df.head()
```

```python
# Basic info
print(df.info())

# Null check
print(df.isnull().sum())

# Unique regions
print(df['Region'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='UnemploymentRate')
plt.title("Unemployment Rate Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot unemployment rate by region
plt.figure(figsize=(14,8))
sns.lineplot(data=df, x='Date', y='UnemploymentRate', hue='Region')
plt.title("Unemployment Rate Over Time by Region")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter data around COVID period
covid_df = df[df['Date'] >= '2020-03-01']
plt.figure(figsize=(14,6))
sns.lineplot(data=covid_df, x='Date', y='UnemploymentRate', hue='Region')
plt.title("COVID-19 Impact on Unemployment (Mar–Nov 2020)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Add 'Month' and 'Year' columns
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year

# Group by Month and get average unemployment rate
monthly_trends = df.groupby('Month')['UnemploymentRate'].mean().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

# Plot monthly trend
plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_trends.index, y=monthly_trends.values, marker='o', color='crimson')
plt.title('Average Unemployment Rate by Month', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
### Insights & Policy Recommendations

- The **highest unemployment rates** were observed in **April and May**, aligning with the national COVID-19 lockdown.
- A noticeable drop begins from **June**, indicating partial recovery.
- **March** shows a rise in unemployment, possibly marking the end of seasonal work or early pandemic effects.
- **September to October** have relatively low unemployment, suggesting post-monsoon or festive season employment peaks.
- Data for **November and December** appears missing, so full-year analysis isn't possible.

#### Policy Suggestions:
- Implement **temporary employment or relief schemes** during April–May to cushion economic shocks.
- Create support for **seasonal workers**, especially around transitional periods like March and July.
- Ensure **year-round data collection**, especially in winter months, to improve future planning.
```
